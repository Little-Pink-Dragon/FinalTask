```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
df = pd.read_csv("/content/kc_house_data.csv", sep=",")
df.head
```

**1.2 Изучите стоимости недвижимости**

```python
plt.figure(figsize=(6, 6))
plt.hist(df['price'])
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество');
```

**1.3 Изучите распределение квадратуры**

```python
plt.figure(figsize=(6, 8))
plt.hist(df['sqft_living'])
plt.title('Распределение жилой квадратуры')
plt.xlabel('Кв. метры жилья')
plt.ylabel('Количество');
```

**1.4 Изучите распределение года постройки**

```python
plt.figure(figsize=(6, 8))
plt.hist(df['yr_built'])
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество');
```

**Условие 3: 3 задача**

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

*На этом графике можно увидеть зависимость, как меняется стомость квартиры от её площади. Чем больше площадь, тем выше стоимость* 

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

*На этом графике можно увидеть, как увеличивается стоимость квартиры с повышением качетсва конструкции и от дизайна комнат*

```python
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('grade')
plt.title('Distribution of price by grade');
```

*Наибольшую стоимость имеют дома с оценкой конструкции и 13 дизайна *

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('floors')
plt.title('Distribution of price by floors');
```

*Набольшую цена у квартир, которых 2,5 комнаты *

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('view')
plt.title('Distribution of price by view');
```

*На этом графике мы видим, что максимальная оценка квартиры - 4 балла. И эти квартры имеют макисмальную стоимость *
